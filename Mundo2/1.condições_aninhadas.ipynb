# 🐍 Aula 12: Condições Aninhadas

## 1. Estrutura de Controle (elif)

Em Python, quando temos múltiplas possibilidades de resposta além do simples "Verdadeiro ou Falso", utilizamos o aninhamento de condições. A palavra-chave `elif` é a junção de `else` + `if`.

| Comando | Descrição | Obrigatoriedade |
| :--- | :--- | :--- |
| `if` | Condição primária para iniciar o bloco. | **Obrigatório** |
| `elif` | Condições intermediárias (pode haver várias). | Opcional |
| `else` | Condição final se nenhuma anterior for atendida. | Opcional |

---

## 2. Exemplo Prático (Aninhamento)

Diferente de uma condição composta simples, aqui podemos tratar vários desvios:

```python
nome = str(input('Qual é o seu nome? '))

if nome == 'Gustavo':
    print('Que nome bonito!')
elif nome == 'Pedro' or nome == 'Maria' or nome == 'Paulo':
    print('Seu nome é bem popular no Brasil.')
elif nome in 'Ana Cláudia Jéssica Juliana':
    print('Belo nome feminino.')
else:
    print('Seu nome é bem normal.')

print(f'Tenha um bom dia, {nome}!')
```
