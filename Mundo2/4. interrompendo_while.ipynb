# 🐍 Aula 15: Interrompendo repetições while

## 1. O Comando `break`

O `break` é utilizado para interromper um laço de repetição imediatamente, jogando a execução para fora do bloco, independentemente da condição do `while`. Isso permite criar "loops infinitos" controlados internamente.

```python
# Loop Infinito com Interrupção Interna
soma = 0
while True:
    n = int(input('Digite um número (999 para parar): '))
    if n == 999:
        break # Interrompe o laço sem somar o 999
    soma += n
print(f'A soma vale {soma}')
```

## 2. F-Strings (Python 3.6+)

As `f-strings` facilitam a exibição de variáveis dentro de textos sem a necessidade do `.format()`. Basta colocar um `f` antes das aspas e o nome da variável entre chaves `{}`.

```python
nome = 'José'
idade = 33
salário = 987.35

# Exemplo de F-String com formatação de casas decimais
print(f'O {nome} tem {idade} anos e ganha R${salário:.2f}.')
```
