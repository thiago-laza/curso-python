# 🐍 Aula 14: Estrutura de Repetição (while)

## 1. O Conceito de Teste Lógico

Diferente do `for` (laço com variável de controle), o `while` é um **laço com teste lógico**. Ele é ideal para situações onde o número de repetições é desconhecido ou depende de uma condição externa.

| Estrutura | Quando usar? | Exemplo prático |
| :--- | :--- | :--- |
| **`for`** | Quando você sabe o limite (intervalo fixo). | Ler 10 notas de alunos. |
| **`while`** | Quando você NÃO sabe o limite (condição de parada). | Ler nomes até que o usuário digite "Sair". |

---

## 2. Funcionamento do `while`

O bloco de código dentro do `while` continuará sendo executado enquanto a condição for **verdadeira**. Assim que a condição se tornar **falsa**, o programa sai do laço.

```python
# Exemplo com limite conhecido (similar ao for)
c = 1
while c < 10:
    print(c)
    c += 1

# Exemplo com limite desconhecido (Flag/Condição de parada)
n = 1
while n != 0:
    n = int(input('Digite um valor (0 para parar): '))
```
