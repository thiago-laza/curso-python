# 🐍 Aula 13: Estrutura de Repetição (for)

## 1. O Conceito de Iteração

Os laços (*loops*), repetições ou **iterações** permitem que um bloco de código seja executado várias vezes sem a necessidade de reescrevê-lo. O `for` é conhecido como um laço com variável de controle.

| Componente | Descrição | Exemplo Python |
| :--- | :--- | :--- |
| **Variável** | O contador que muda a cada volta. | `for c in ...` |
| **Intervalo** | Define o início, fim e o passo. | `range(0, 10, 1)` |
| **Bloco** | Comandos que serão repetidos. | (Recuados com Tab) |

---

## 2. Funcionamento do `range()`

O segredo do `for` em Python está na função `range(início, fim, passo)`. 
**Importante:** O Python sempre ignora o último número do intervalo.

```python
# Contagem progressiva (1 a 6)
for c in range(1, 7):
    print(c)

# Contagem regressiva (10 a 0)
for c in range(10, -1, -1):
    print(c)

# Pulando de 2 em 2
for c in range(0, 10, 2):
    print(c)
```
