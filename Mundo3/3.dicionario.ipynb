# 🐍 Aula 19: Dicionários

## 1. Estrutura de um Dicionário
Um dicionário é composto por pares de **Chave: Valor**. 


* **Chaves (Keys):** São os identificadores (ex: 'nome', 'idade').
* **Valores (Values):** São os dados armazenados (ex: 'Gustavo', 22).
* **Itens (Items):** É o conjunto composto pela Chave e pelo Valor.

### **Comandos Básicos:**
```python
dados = {'nome': 'Pedro', 'idade': 25}

print(dados['nome'])  # Saída: Pedro
dados['sexo'] = 'M'   # Adiciona um novo elemento (não precisa de .append)
del dados['idade']    # Remove um elemento
```

---

## 2. Métodos Úteis e Laços
Diferente das listas, para percorrer um dicionário com o `for`, usamos métodos específicos:

```python
filme = {'titulo': 'Star Wars', 'ano': 1977, 'diretor': 'George Lucas'}

print(filme.values()) # Retorna: Star Wars, 1977...
print(filme.keys())   # Retorna: titulo, ano, diretor
print(filme.items())  # Retorna os pares completos

for k, v in filme.items():
    print(f'O {k} é {v}')
```
