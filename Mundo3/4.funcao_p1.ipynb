# 🐍 Aula 20: Funções (Parte 1)

## 1. O que são Funções (`def`)?
Funções são trechos de código que realizam tarefas específicas e podem ser chamados a qualquer momento no seu programa. Elas servem para automatizar **rotinas**.

* **Sintaxe Básica:**
    ```python
    def mostrar_linha():
        print('-' * 30)

    # Chamada da função
    mostrar_linha()
    print('   CURSO EM VÍDEO   ')
    mostrar_linha()
    ```

## 2. Funções com Parâmetros
Você pode passar informações para dentro de uma função para que ela execute tarefas personalizadas.

```python
def mensagem(msg):
    print('-' * 30)
    print(msg)
    print('-' * 30)

mensagem('SISTEMA DE ALUNOS')
```

## 3. Empacotamento de Parâmetros (`*args`)
Quando você não sabe quantos valores o usuário vai passar, você pode "empacotar" os parâmetros usando o símbolo de asterisco `*`. O Python criará uma **Tupla** com esses valores.

```python
def contador(*num):
    tamanho = len(num)
    print(f'Recebi os valores {num} e são ao todo {tamanho} números.')

contador(2, 1, 7)
contador(8, 0)
```
