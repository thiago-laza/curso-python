# 🐍 Aula 17: Listas (Parte 2) - Listas Compostas

## 1. O que são Listas Compostas?
Uma lista composta é simplesmente uma lista que contém outras listas como seus elementos. Isso permite organizar dados relacionados, como o nome e a idade de várias pessoas, em uma única estrutura.

### **Cópia de Listas (Importante!):**
Ao inserir uma lista dentro de outra, use o fatiamento `[:]` para criar uma **cópia**. Se você não fizer isso, o Python criará uma **ligação** entre elas, e alterar a lista original mudará os dados dentro da lista composta também.

```python
dados = ['Pedro', 25]
pessoas = list()
# Criando uma cópia e adicionando à lista composta
pessoas.append(dados[:]) 
```

---

## 2. Acessando Dados em Listas Compostas
Para acessar um dado específico, usamos dois índices: o primeiro para a lista interna e o segundo para o elemento dentro dela.

```python
pessoas = [['Pedro', 25], ['Maria', 19], ['João', 32]]

print(pessoas[0][0]) # Saída: Pedro
print(pessoas[1][1]) # Saída: 19
print(pessoas[2][0]) # Saída: João
```
