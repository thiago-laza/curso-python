# 🐍 Aula 23: Tratamento de Erros e Exceções

## 1. Erros vs. Exceções
* **Erro de Sintaxe:** Ocorre quando você escreve o código errado (ex: `primt` em vez de `print`). O programa nem começa a rodar.
* **Exceção:** O código está escrito corretamente, mas algo inesperado acontece durante a execução (ex: dividir por zero, digitar texto onde se esperava um número).


---

## 2. A Estrutura `try... except`
Para tratar exceções, usamos um bloco de controle especial:

| Bloco | Função |
| :--- | :--- |
| **`try:`** | Tenta executar esse código. |
| **`except:`** | O que fazer se der erro. |
| **`else:`** | O que fazer se der CERTO (opcional). |
| **`finally:`** | O que fazer independente se deu certo ou errado (opcional). |

```python
try:
    a = int(input('Numerador: '))
    b = int(input('Denominador: '))
    r = a / b
except (ValueError, TypeError):
    print('Tivemos um problema com os tipos de dados que você digitou.')
except ZeroDivisionError:
    print('Não é possível dividir um número por zero!')
else:
    print(f'O resultado é {r}')
finally:
    print('Volte sempre!')
```
