# 🐍 Aula 16: Variáveis Compostas (Tuplas)

## 1. O que são Tuplas?
Até agora, você usou variáveis simples que guardam apenas um valor por vez. As **Tuplas** permitem armazenar vários valores em uma única estrutura, acessíveis através de índices (0, 1, 2...).

### **A Regra de Ouro:**
> **AS TUPLAS SÃO IMUTÁVEIS.** > Isso significa que, uma vez definida, você não pode alterar, adicionar ou remover itens de uma tupla enquanto o programa estiver em execução.

| Característica | Detalhe |
| :--- | :--- |
| **Sintaxe** | Definidas entre parênteses `()` (opcional em versões novas, mas recomendado). |
| **Acesso** | Utiliza colchetes `variável[índice]`. |
| **Fatiamento** | Aceita as mesmas regras de strings: `[1:3]`, `[:2]`, `[-1]`, etc. |


---

## 2. Manipulando Tuplas com `for`
Existem três formas principais de percorrer uma tupla:

1.  **Forma Simples (Só o valor):**
    ```python
    for comida in lanche:
        print(f'Vou comer {comida}')
    ```
2.  **Usando `range` (Valor e Posição):**
    ```python
    for cont in range(0, len(lanche)):
        print(f'Vou comer {lanche[cont]} na posição {cont}')
    ```
3.  **Usando `enumerate` (Valor e Posição - mais moderno):**
    ```python
    for pos, comida in enumerate(lanche):
        print(f'Vou comer {comida} na posição {pos}')
    ```
