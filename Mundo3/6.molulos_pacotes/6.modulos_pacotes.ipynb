# 🐍 Aula 22: Módulos e Pacotes

## 1. Modularização
A modularização surgiu na década de 60 com o objetivo de dividir programas grandes em pedaços menores (módulos), aumentando a legibilidade e facilitando a manutenção.

* **Módulo:** Em Python, todo arquivo `.py` pode ser considerado um módulo.
* **Como usar:** Basta importar o arquivo usando `import nome_do_arquivo` ou `from nome_do_arquivo import funcao`.


---

## 2. Pacotes
Quando um projeto cresce e tem muitos módulos, nós os organizamos em **Pacotes**. Um pacote é basicamente uma **pasta** que contém módulos, organizada por assuntos.

* **Estrutura de Pastas:**
    * `utilidades/` (Pasta do Pacote)
        * `__init__.py` (Arquivo obrigatório para o Python reconhecer a pasta como pacote)
        * `moeda.py` (Módulo)
        * `dados.py` (Módulo)
