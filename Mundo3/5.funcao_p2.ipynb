# 🐍 Aula 21: Funções (Parte 2)

## 1. Interactive Help (`help()`)
O Python possui um sistema de ajuda integrado. Você pode usar a função `help()` no console para ver o manual de qualquer comando ou biblioteca.
*Ex: `help(print)` ou `help(len)`.*

## 2. Docstrings
São strings de documentação que você cria para as suas próprias funções. Elas devem ser colocadas logo abaixo da linha `def`, entre três aspas duplas `"""`.
```python
def contador(i, f, p):
    """
    -> Faz uma contagem e mostra na tela.
    :param i: início da contagem
    :param f: fim da contagem
    :param p: passo da contagem
    :return: sem retorno
    """
    # código aqui...
```

## 3. Parâmetros Opcionais
Você pode definir valores padrão para os parâmetros. Se o usuário não informar o valor, o Python usará o padrão.
```python
def somar(a=0, b=0, c=0):
    s = a + b + c
    print(f'A soma vale {s}')

somar(3, 2) # c passará a valer 0 automaticamente
```

## 4. Escopo de Variáveis
* **Escopo Global:** Variáveis declaradas no programa principal valem em todo o código.
* **Escopo Local:** Variáveis declaradas dentro de uma função valem apenas ali.
* **Palavra `global`:** Use para manipular uma variável global dentro de uma função sem criar uma versão local dela.

## 5. Retorno de Valores (`return`)
As funções podem retornar um resultado para o programa principal, permitindo maior personalização da exibição.
```python
def somar(a=0, b=0, c=0):
    s = a + b + c
    return s

r1 = somar(3, 2, 5) # r1 recebe o valor 10
```
