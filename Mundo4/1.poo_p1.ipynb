# 🐍 Mundo 4: Programação Orientada a Objetos (POO)

## 1. Conceitos Fundamentais
A POO tenta aproximar a programação do mundo real, criando moldes para gerar objetos com características e comportamentos próprios.

* **Classe:** É o molde ou a planta (ex: Forminha de biscoito, Planta de uma casa).
* **Objeto:** É o item real criado a partir da classe (ex: O biscoito, A casa construída).
* **Atributos:** São as características do objeto (dados/variáveis).
* **Métodos:** São as ações que o objeto pode realizar (funções).
* **Estado:** É o conjunto de valores dos atributos num determinado momento.
* **Instanciar:** O ato de criar um objeto a partir de uma classe.


---

## 2. Estrutura Básica em Python
```python
class Gafanhoto:
    # Método Construtor (Inicialização)
    def __init__(self, nome, idade):
        self.nome = nome     # Atributo de instância
        self.idade = idade   # Atributo de instância

    # Método de Instância (Ação)
    def aniversario(self):
        self.idade += 1

# Instanciação (Criação de Objeto)
g1 = Gafanhoto("Thiago", 28)
g1.aniversario()
print(f"{g1.nome} agora tem {g1.idade} anos.")
```

---

## 3. Métodos e Atributos "Mágicos" (Dunders)
O Python usa nomes com duplo *underscore* (`__`) para funcionalidades especiais:
* `__init__`: Construtor (executado ao criar o objeto).
* `__doc__`: Acede à documentação (*docstring*) da classe.
* `__str__`: Define o que aparece quando dás um `print(objeto)`.
* `__dict__` ou `__getstate__`: Mostra o estado do objeto em formato de dicionário.
