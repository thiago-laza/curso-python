# 🐍 Aula 06: Tipos Primitivos e Saída de Dados

## 1. Os 4 Tipos Primitivos Básicos

Em Python, toda variável pertence a um tipo. Se você não especificar, o `input()` lerá tudo como **String**.

| Tipo | Nome | Descrição | Exemplos |
| --- | --- | --- | --- |
| **`int`** | Inteiro | Números sem casas decimais (positivos ou negativos) | `7`, `-4`, `0`, `9875` |
| **`float`** | Real | Números com ponto flutuante (decimais) | `4.5`, `0.076`, `-15.2`, `7.0` |
| **`bool`** | Booleano | Valores lógicos (sempre inicia com Maiúscula) | `True`, `False` |
| **`str`** | String | Conjunto de caracteres (sempre entre aspas) | `'Olá'`, `'7.5'`, `' '` |

---

## 2. Conversão de Dados (Casting)

Para realizar cálculos com dados vindos do teclado, é necessário converter o valor:

```python
n1 = int(input('Digite um número: '))      # Converte para Inteiro
n2 = float(input('Digite um peso: '))     # Converte para Real
n3 = bool(input('Digitou algo? '))        # Retorna True se houver valor, False se vazio

```

---

## 3. Saída de Dados (Método `.format()` e f-string)

A forma moderna de exibir mensagens combinando texto e variáveis:

```python
n1 = int(input('Valor 1: '))
n2 = int(input('Valor 2: '))
s = n1 + n2

# Usando chaves {} como máscaras
print('A soma entre {} e {} vale {}'.format(n1, n2, s))
print(f'A soma entre {n1} e {n2} vale {s}')

```

---

## 4. Métodos de Teste de Tipo (`is...`)

Utilizados para investigar a natureza de uma variável (retornam sempre `True` ou `False`). Úteis para validação antes da conversão.

* **`.isnumeric()`**: É um número?
* **`.isalpha()`**: É alfabético (letras)?
* **`.isalnum()`**: É alfanumérico (letras ou números)?
* **`.isupper()`**: Está tudo em maiúsculas?
* **`.islower()`**: Está tudo em minúsculas?
* **`.isspace()`**: Contém apenas espaços?
* **`.istitle()`**: Está capitalizado (ex: 'Olá Mundo')?

---

## 5. Exemplo de Código para Estudo

Copie e execute no seu notebook para ver os testes em ação:

```python
algo = input('Digite algo: ')
print('O tipo primitivo desse valor é:', type(algo))
print('É um número?', algo.isnumeric())
print('É alfabético?', algo.isalpha())
print('É alfanumérico?', algo.isalnum())
print('Está em maiúsculas?', algo.isupper())
print('Está em minúsculas?', algo.islower())
print('Está capitalizada?', algo.istitle())

```

### Exercício 01 (Desafio 03 do Curso)

**Objetivo:** Praticar a conversão de tipos (casting) e operações aritméticas simples.

* **Enunciado:** Crie um programa em Python que leia dois números inteiros digitados pelo usuário. O programa deve calcular a soma entre esses dois números e exibir o resultado na tela com uma mensagem personalizada.
* **Requisito técnico:** Utilize o método `.format()` para exibir a saída (ex: `A soma entre {} e {} vale {}`).
* **Dica:** Lembre-se de usar a função `int()` para envolver o `input()`, caso contrário o Python tratará os números como texto e irá apenas juntá-los (ex: 5 + 2 viraria 52).









````python
n1 = int(input('Digite um numero: '))
n2 = int(input('Digite outro numero: '))

s = n1 + n2

print(f'A soma entre {n1} e {n2} e igual a {s}')
````

### Exercício 02 (Desafio 04 do Curso)

**Objetivo:** Explorar os métodos de teste de tipo (objetos da classe String) e a análise de dados.

* **Enunciado:** Faça um programa que leia algo pelo teclado (qualquer frase, número ou caractere) e mostre na tela o seu tipo primitivo. Além disso, o programa deve exibir todas as informações possíveis sobre o que foi digitado.
* **Requisito técnico:** Utilize os métodos de teste `is...` para informar:
1. É somente espaços? (`.isspace()`)
2. É um número? (`.isnumeric()`)
3. É alfabético? (`.isalpha()`)
4. É alfanumérico? (`.isalnum()`)
5. Está em maiúsculas? (`.isupper()`)
6. Está em minúsculas? (`.islower()`)
7. Está capitalizado (primeira letra maiúscula)? (`.istitle()`)

````python
entrada = input('Digite qualquer coisa: ')

print(f'O tipo primitivo de {entrada} e {type(entrada)}')

print(f'So tem espacos? {entrada.isspace()}')
print(f'E um numero? {entrada.isnumeric()}')
print(f'E alfabetico? {entrada.isalpha()}')
print(f'E alfanumerico? {entrada.isalnum()}')
print(f'Esta em maiusculas? {entrada.isupper()}')
print(f'Esta em minusculas? {entrada.islower()}')
print(f'Esta capitalizada? {entrada.istitle()}')
````
