# 🐍 Aula 10: Condições (Parte 1)

Nesta aula, aprendemos a dar "poder de decisão" ao programa através de estruturas condicionais simples e compostas.

## 1. Conceito de Condição

Uma condição permite que o programa siga caminhos diferentes dependendo de um resultado lógico (Verdadeiro ou Falso).

* **Estrutura Simples**: Executa um bloco de código apenas se a condição for verdadeira.
* **Estrutura Composta**: Oferece um caminho alternativo (`else`) caso a condição seja falsa.

---

## 2. Sintaxe em Python

A indentação (espaçamento) é obrigatória para definir o que pertence ou não à condição.

```python
if condição:
    # bloco executado se for Verdadeiro
else:
    # bloco executado se for Falso

```

### Exemplo Prático:

```python
tempo = int(input('Quantos anos tem seu carro? '))
if tempo <= 3:
    print('Carro novo')
else:
    print('Carro velho')
print('--FIM--') # Sempre será executado, pois está fora da indentação

```

---

## 3. Condição Simplificada

Python permite escrever condições simples em apenas uma linha (operador ternário):

```python
tempo = int(input('Quantos anos tem seu carro? '))
print('carro novo' if tempo <= 3 else 'carro velho')

```

---

## 4. Exemplos de Uso Comum

* **Média de Aluno**:

```python
n1 = float(input('Primeira nota: '))
n2 = float(input('Segunda nota: '))
m = (n1 + n2) / 2
print(f'Sua média foi {m:.1f}')
if m >= 6.0:
    print('Sua média foi boa! PARABÉNS!')
else:
    print('Sua média foi baixa! ESTUDE MAIS!')

```

* [ ] **Desafio 28: Jogo da Adivinhação v.1.0**
* Escreva um programa que faça o computador "pensar" em um número inteiro entre 0 e 5 e peça para o usuário tentar descobrir qual foi o número escolhido pelo computador. O programa deverá escrever na tela se o usuário venceu ou perdeu.




````python
from random import randint

valor = randint(0,5)

numero = int(input('Informe um numero inteiro de 0 a 5: '))

if numero == valor:
    print(f'Voce venceu, o numero {valor} foi o que o computador pensou.')
else:
    print(f'Voce perdeu, o computador pensou no numero: {valor}')

````

* [ ] **Desafio 29: Radar eletrônico**
* Escreva um programa que leia a velocidade de um carro. Se ele ultrapassar 80Km/h, mostre uma mensagem dizendo que ele foi multado. A multa vai custar R$7,00 por cada Km acima do limite.




````python
velocidade = float(input('Velocidade do carro: '))

if velocidade <= 80:
    print(f'Velocidade de {velocidade} km/h, dentro do limite permitido.')
else:
    multa = (velocidade - 80) * 7
    print(f'Velocidade: {velocidade} km/h, multado em R$ {multa:.2f}')
````

* [ ] **Desafio 31: Custo da Viagem**
* Desenvolva um programa que pergunte a distância de uma viagem em Km. Calcule o preço da passagem, cobrando R0,50 por Km para viagens de até 200Km e R0,45 para viagens mais longas.




````python
distancia = float(input('Informe a distancia em km: '))

if distancia <= 200:
    valor = distancia * 0.5
    print(f'Valor da viagem: R$ {valor:.2f}')
else:
    valor = distancia * 0.45
    print(f'Valor da viagem: R$ {valor:.2f}')
````

* Faça um programa que leia um ano qualquer e mostre se ele é BISSEXTO.




````python
ano = int(input('Informe o ano: '))

if (ano % 4 == 0) and (ano % 100 != 0) or (ano % 400 == 0):
    print(f'{ano} e bissexto')
else:
    print(f'{ano} nao e bissexto')
````

* [ ] **Desafio 33: Maior e menor valores**
* Faça um programa que leia três números e mostre qual é o MAIOR e qual é o MENOR.




````python
n1 = float(input('Informe o primeiro numero: '))
n2 = float(input('Informe o segundo numero: '))
n3 = float(input('Informe o terceiro numero: '))

menor = n1

if n2 < n1 and n2 < n3:
    menor = n2
if n3 < n1 and n3 < n2:
    menor = n3
    
maior = n1

if n2 > n1 and n2 > n3:
    maior = n2
if n3 > n1 and n3 > n2:
    maior = n3
    
print(f'Menor: {menor}')
print(f'Maior: {maior}')
````

* [ ] **Desafio 34: Aumentos múltiplos**
* Escreva um programa que pergunte o salário de um funcionário e calcule o valor do seu aumento. Para salários superiores a R$1.250,00, calcule um aumento de 10%. Para os inferiores ou iguais, o aumento é de 15%.




````python
salario = float(input('Informe o salario: R$ '))

if salario <= 1250:
    novo_salario = salario * 1.15
    print(f'Novo salario (+15%): {novo_salario:.2f}')
else:
    novo_salario = salario * 1.1
    print(f'Novo salario (+10%): {novo_salario:.2f}')
````

* [ ] **Desafio 35: Analisando Triângulo v1.0**
* Desenvolva um programa que leia o comprimento de três retas e diga ao usuário se elas podem ou não formar um triângulo.

````python
l1 = float(input('Informe a medida do primeiro segmento: '))
l2 = float(input('Informe a medida do segundo segmento: '))
l3 = float(input('Informe a medida do terceiro lado: '))

if l1 + l2 > l3 and l1 + l3 > l2 and l2 + l3 > l1:
    print(f'{l1}, {l2} e {l3} podem formar um triangulo')
else:
    print(f'{l1}, {l2} e {l3} nao podem formar um triangulo')
````
