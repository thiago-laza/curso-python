# 🐍 Aula 08: Utilizando Módulos

Nesta aula, aprendemos como expandir as funcionalidades básicas do Python importando bibliotecas (módulos) que trazem novos comandos e ferramentas para os nossos programas.

## 1. O que são Módulos?

Imagine que o Python "padrão" é como um carro básico. Ele funciona bem, mas você pode instalar "módulos" adicionais, como ar-condicionado ou vidros elétricos, para ganhar novas funções.

* O Python mantém apenas os comandos essenciais carregados por padrão para ser rápido e economizar memória.
* Quando precisamos de algo extra, fazemos a **importação**.

---

## 2. Formas de Importação

Existem duas maneiras principais de trazer funcionalidades de um módulo para seu código:

### A. Importação Geral (`import`)

Importa **todas** as funcionalidades do módulo.

* **Comando:** `import nome_do_modulo`
* **Exemplo:** `import math`
* **Uso:** Para chamar uma função, você deve usar o prefixo: `math.sqrt(num)`.

### B. Importação Específica (`from ... import`)

Importa apenas as funcionalidades **escolhidas**, economizando memória.

* **Comando:** `from nome_do_modulo import funcao`
* **Exemplo:** `from math import sqrt, floor`
* **Uso:** Você chama a função diretamente: `sqrt(num)`.

---

## 3. A Biblioteca `math` (Matemática)

Uma das bibliotecas padrão mais utilizadas. Algumas de suas principais funções são:

| Função | O que faz |
| --- | --- |
| `ceil` | Arredonda um número para **cima**. |
| `floor` | Arredonda um número para **baixo**. |
| `trunc` | Elimina os números após a vírgula (trunca), sem arredondar. |
| `pow` | Calcula a potência (semelhante ao operador `**`). |
| `sqrt` | Calcula a **raiz quadrada** (*square root*). |
| `factorial` | Calcula o fatorial de um número. |

---

## 4. Outros Módulos Úteis mencionados

* **`random`**: Usado para gerar números aleatórios. Ex: `random.randint(1, 10)` gera um número inteiro entre 1 e 10.
* **`emoji`**: Um módulo externo (precisa ser instalado) que permite usar emojis no terminal.

---

## 5. Como instalar Módulos Externos (PyPI)

Muitos módulos são criados pela comunidade e não vêm no Python. Eles ficam no **PyPI** (*Python Package Index*).
No **PyCharm**, você pode instalar assim:

1. Vá em `File` > `Settings` (ou `Preferences` no Mac).
2. Vá em `Project` > `Python Interpreter`.
3. Clique no botão **`+`**, pesquise o módulo (ex: `emoji`) e clique em **`Install Package`**.

* [ ] **Desafio 17: Catetos e Hipotenusa**
* Leia o comprimento do cateto oposto e do cateto adjacente de um triângulo retângulo, calcule e mostre o comprimento da hipotenusa.




```python
from math import sqrt

cat1 = float(input('Informe a medida do primeiro cateto: '))
cat2 = float(input('Informe a medida do segundo cateto: '))

hip = sqrt(cat1**2 + cat2**2 )

print(f'Um triangulo retangulo, cujos catetos medem {cat1} e {cat2} tem hipotenusa medindo {hip}')
```

* [ ] **Desafio 18: Seno, Cosseno e Tangente**
* Leia um ângulo qualquer e mostre o valor do seno, cosseno e tangente desse ângulo.




```python
from math import sin, cos, tan, radians

angulo = float(input('Informe a medida do angulo em grau: '))
rad = radians(angulo)

seno = sin(rad)
cosseno = cos(rad)
tangente = tan(rad)

print(f'O angulo com medida de {angulo}⁰, tem sen({angulo}⁰) = {seno:.2f}, cos({angulo}⁰) = {cosseno:.2f} e tg({angulo}) = {tangente:.2f}')
```

* [ ] **Desafio 19: Sorteando um item na lista**
* Um professor quer sortear um dos seus quatro alunos para apagar o quadro. Faça um programa que leia o nome deles e escreva o nome do escolhido.




```python
from random import choice

aluno1 = input('Escreva o nome do primeiro aluno: ')
aluno2 = input('Escreva o nome do segundo aluno: ')
aluno3 = input('Escreva o nome do terceiro aluno: ')
aluno4 = input('Escreva o nome do quarto aluno: ')

lista = [aluno1, aluno2, aluno3, aluno4]

print(choice(lista))
```

* [ ] **Desafio 20: Sorteando uma ordem na lista**
* O mesmo professor quer sortear a ordem de apresentação de trabalhos dos alunos. Leia o nome dos quatro alunos e mostre a ordem sorteada.




```python
from random import shuffle

aluno1 = input('Escreva o nome do primeiro aluno: ')
aluno2 = input('Escreva o nome do segundo aluno: ')
aluno3 = input('Escreva o nome do terceiro aluno: ')
aluno4 = input('Escreva o nome do quarto aluno: ')

lista = [aluno1, aluno2, aluno3, aluno4]
shuffle(lista)

print(lista)
```

* [ ] **Desafio 21: Tocando um MP3**
* Faça um programa em Python que abra e reproduza o áudio de um arquivo MP3.

```python
import pygame

# Inicialização padrão
pygame.mixer.init()
pygame.mixer.music.load('pitch2.mp3')
pygame.mixer.music.play()

# Criamos uma variável para receber o comando do usuário
# Usamos o .lower() para aceitar tanto 'p' quanto 'P'
comando = input('Música tocando... Digite "P" para parar: ').lower()

# Estrutura Condicional da Aula 10
if comando == 'p':
    pygame.mixer.music.stop()
    print('Áudio interrompido pelo usuário.')
else:
    print('Você não digitou "P", mas o programa encerrou e a música parou.')
```
