# 🐍 Aula 07: Operadores Aritméticos em Python

## 1. Tabela de Operadores

| Operador | Operação | Exemplo | Resultado |
| --- | --- | --- | --- |
| `+` | Adição | `5 + 2` | `7` |
| `-` | Subtração | `5 - 2` | `3` |
| `*` | Multiplicação | `5 * 2` | `10` |
| `/` | Divisão (Real) | `5 / 2` | `2.5` |
| `**` | Potência | `5 ** 2` | `25` |
| `//` | Divisão Inteira | `5 // 2` | `2` |
| `%` | Módulo (Resto) | `5 % 2` | `1` |

---

## 2. Ordem de Precedência

Na matemática e no Python, resolvemos as expressões nesta ordem:

1. **`()`** Parênteses
2. **`**`** Potências
3. **`*`**, **`/`**, **`//`**, **`%`** (Multiplicação, divisões e resto)
4. **`+`**, **`-`** (Soma e subtração)

> **Nota:** Se houver operadores de mesma prioridade (como `*` e `/`), resolvemos quem aparecer primeiro da **esquerda para a direita**.

---

## 3. Truques e Atalhos

* **Raiz Quadrada:** `n ** (1/2)` ou `n ** 0.5`
* **Raiz Cúbica:** `n ** (1/3)`
* **Repetir String:** `'A' * 10` (Gera `AAAAAAAAAA`)
* **Concatenar:** `'Oi' + 'Ana'` (Gera `OiAna`)

---

## 4. Formatação de Saída (`print`)

### Alinhamento de Texto

```python
nome = 'Python'
print(f'|{nome:20}|')    # 20 espaços
print(f'|{nome:>20}|')   # Alinhado à direita
print(f'|{nome:<20}|')   # Alinhado à esquerda
print(f'|{nome:^20}|')   # Centralizado
print(f'|{nome:=^20}|')  # Centralizado com preenchimento

```

### Casas Decimais e Quebra de Linha

```python
div = 10 / 3
# Exibir com 2 casas decimais
print(f'O resultado é {div:.2f}') 

# Não pular linha ao final do print
print('Olá', end=' ')
print('Mundo')

# Quebrar linha manualmente
print('Primeira linha\nSegunda linha')

```

* [ ] **Desafio 05: Antecessor e Sucessor**
* **Enunciado:** Faça um programa que leia um número inteiro e mostre na tela o seu sucessor e seu antecessor.
* *Exemplo de saída:* `O antecessor de 5 é 4 e o sucessor é 6.`










````python
n = int(input('Digite um numero: '))

antecessor = n - 1
sucessor = n + 1

print(f'O antecessor de {n} e {antecessor} e o sucessor de {n} e {sucessor}')
````

* [ ] **Desafio 06: Dobro, Triplo, Raiz Quadrada**
* **Enunciado:** Crie um algoritmo que leia um número e mostre o seu dobro, triplo e raiz quadrada.
* *Dica:* Raiz quadrada pode ser feita com `n ** (1/2)`.



````python
n = int(input('Digite um numero: '))

dobro = 2 * n
triplo = 3 * n
raiz_quadrada = n ** (0.5)

print(f'O dobro de {n} e {dobro}')
print(f'O triplo de {n} e {triplo}')
print(f'A raiz quadrada de {n} e {raiz_quadrada}')
````

* [ ] **Desafio 07: Média Aritmética**
* **Enunciado:** Desenvolva um programa que leia as duas notas de um aluno, calcule e mostre a sua média.
* *Atenção:* Cuidado com a ordem de precedência na soma das notas!




````python
n1 = float(input('Digite a primeira nota: '))
n2 = float(input('Digite a segunda nota: '))

media = (n1 + n2) / 2

print(f'A media aritmetica das notas {n1} e {n2} e {media}')
````

* [ ] **Desafio 09: Tabuada**
* **Enunciado:** Faça um programa que leia um número inteiro qualquer e mostre na tela a sua tabuada.
* *Estética:* Tente usar a formatação `f'{n} x {1:2} = {n*1}'` para alinhar os resultados.

* [ ] **Desafio 14: Conversor de Temperaturas**
* **Enunciado:** Escreva um programa que converta uma temperatura digitada em **°C** (Graus Celsius) e a converta para **°F** (Graus Fahrenheit).
* **Fórmula:** A fórmula de conversão é:
$$F = \frac{9 \times C}{5} + 32$$
* *Dica:* Lembre-se da ordem de precedência. No Python, você pode escrever: `((9 * c) / 5) + 32` ou simplesmente `9 * c / 5 + 32`.







````python
temperatura_c = float(input('Informe a temperatura em ⁰C' ))

temperatura_f = ((9 * temperatura_c) / 5) + 32

print(f'A temperatura de {temperatura_c} ⁰C equivale a {temperatura_f} ⁰F')
````

* [ ] **Desafio 15: Aluguel de Carros**
* **Enunciado:** Escreva um programa que pergunte a quantidade de **Km percorridos** por um carro alugado e a quantidade de **dias** pelos quais ele foi alugado. Calcule o preço a pagar, sabendo que o carro custa **R$ 60,00 por dia** e **R$ 0,15 por Km** rodado.
* **Cálculo:** `Total = (Dias * 60) + (Km * 0.15)`

````python
distancia = float(input('Informe a distancia percorrida em Km: '))
dias = int(input('Informe a quantidade de dias com o carro: '))

valor = (60 * dias) + (0.15 * distancia)

print(f'O aluguel de um carro por {dias} dias e {distancia} km percorrido custara R$ {valor:.2f}')
````
