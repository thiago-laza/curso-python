# 🐍 Aula 09: Manipulando Texto

Nesta aula, aprendemos a tratar **Cadeias de Caracteres** (Strings). No Python, toda string é tratada como um objeto e seus caracteres são indexados, começando do **índice 0**.

## 1. Fatiamento de String

O fatiamento permite extrair pedaços de uma string usando colchetes `[]`.

| Comando | O que faz |
| --- | --- |
| `frase[9]` | Seleciona o caractere no índice 9. |
| `frase[9:13]` | Pega do índice 9 até o 12 (o 13 é excluído). |
| `frase[9:21:2]` | Pega do 9 ao 20, pulando de 2 em 2. |
| `frase[:5]` | Pega do início até o índice 4. |
| `frase[15:]` | Pega do índice 15 até o final. |
| `frase[9::3]` | Começa no 9, vai até o final e pula de 3 em 3. |

---

## 2. Análise de String

Ferramentas para descobrir informações sobre o texto.

* **`len(frase)`**: Retorna o comprimento (quantidade de caracteres) da string.
* **`frase.count('o')`**: Conta quantas vezes a letra 'o' (minúscula) aparece.
* **`frase.find('deo')`**: Indica em qual índice começou a sequência 'deo'. Se não encontrar, retorna `-1`.
* **`'Curso' in frase`**: Verifica se a palavra 'Curso' existe na frase (retorna `True` ou `False`).

---

## 3. Transformação de String

Embora as strings sejam **imutáveis** (você não pode alterar um caractere individual diretamente), você pode criar novas versões delas com métodos.

| Método | O que faz |
| --- | --- |
| `replace('Python', 'Android')` | Procura por 'Python' e substitui por 'Android'. |
| `upper()` | Transforma tudo em **MAIÚSCULAS**. |
| `lower()` | Transforma tudo em **minúsculas**. |
| `capitalize()` | Só o primeiro caractere da frase fica em maiúscula. |
| `title()` | Capitaliza a primeira letra de **cada palavra**. |
| `strip()` | Remove espaços inúteis no início e no fim. |
| `lstrip()` / `rstrip()` | Remove espaços apenas da esquerda (*left*) ou direita (*right*). |

---

## 4. Divisão e Junção

* **`frase.split()`**: Divide a string em uma **lista**, usando os espaços como separador por padrão.
* **`'-'.join(frase)`**: Junta os elementos de uma lista usando o separador '-' para formar uma string única.

---

print("""Este é um texto

9. Letra Central: Leia uma palavra de tamanho ímpar e exiba qual é a letra que ocupa a posição central exata.

````python
palavra = input('Digite uma palavra: ')

tamanho = len(palavra)

if tamanho % 2 == 0:
    print(f'"{palavra}" nao tem uma letra central')
else:
    valor_posicao = (tamanho // 2) 
    letra_central = palavra[valor_posicao]
    print(f'Na palavra "{palavra}" a posicao central esta a letra "{letra_central}"')
````

12. Verificador de Extensão: Leia o nome de um arquivo (ex: foto.jpg) e diga se ele é uma imagem (.jpg ou .png) ou um documento (.pdf).

````python
arquivo = input('Arquivo: ')

posicao = arquivo.find('.')

verificador = arquivo[posicao:]

if verificador == '.jpg' or verificador == '.png':
    print(f'O arquivo {arquivo} e uma imagem.')
elif verificador == '.pdf':
    print(f'O arquivo {arquivo} e um documento')
else:
    print(f'O arquivo {arquivo} tem uma extensao nao conhecida.')
````
